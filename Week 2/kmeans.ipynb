```python
# 24 people have null incomes
# handling null values by taking mean of existing incomes and replacing them with null

customer["Income"] = customer["Income"].fillna(customer["Income"].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
customer.columns
```

```python
# creating a new column/feature which is Total_Spend.
# Total_Spend will be the sum of all amounts spent on all commodities.

customer['Total_Spend'] = (customer["MntFishProducts"] + customer["MntFruits"] + customer["MntWines"] + customer["MntMeatProducts"]  + customer["MntSweetProducts"] + customer["MntGoldProds"] )

customer['Total_Spend'].head()

X = customer[["Income", "Total_Spend"]]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#configuring of points

from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)

customer["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
customer[["Income", "Total_Spend", "Cluster"]].head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_scaled[:,0],
    X_scaled[:,1],
    c=customer["Cluster"],
    cmap="viridis",
    s=50
)

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    c="red",
    marker="X",
    s=200,
    label="Centroids"
)

plt.xlabel("Income (Scaled)")
plt.ylabel("Total Spent (Scaled)")
plt.title("Customer Segmentation using K-Means")
plt.legend()
plt.show()
```
